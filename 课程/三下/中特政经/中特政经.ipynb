## 一社会主义经济制度的建立 ##
我国社会主义经济制度的建立主要通过以下途径实现
1. 国营经济的形成和壮大
2. 生产资料私有制的社会主义改造



五、国有经济的主导作用
1. 生产关系方面
- 在多种所有制结构中，国有经济起支配地位，保证所有制经济的社会主义方向

2. 生产力方面
- 体现在国民经济中的控制力，能够保证国民经济持续协调健康发展
- 第一，做私人资本不愿意做、不能做的事
- 第二，国家进行社会调节的重要载体和依托力量
- 第三，保证公共产品和服务的提供
- 第四，保证国家战略性资源、力量

六、做强做优做大国有资本和国有企业（顺序别乱）
- 国有经济的主导作用是通过国有企业和国有资本实现的
- 国有企业和一般企业的异同
- 国有资本和一般资本的异同
- 做强做优做大国有企业与做强做优做大国有资本的关系


2. 农村经营体制改革
- 改革开放后农村基本经营制度：以家庭承包为基础、统分结合的双层经营体质

3. 新时代农村基本经营制度改革方向
- 第一，坚持农村土地集体所有制不变
- 第二，稳定承包经营权不变，推进“三权分置”改革
- 第三，坚持最严格的耕地保护制度，赋予农民土地更多权利
- 第四，发展农村新型合作组织
- 第五，稳步推进农村集体产权制度改革

4. 非公有制经济——同公有制经济相对应生产资料归私人所有，属于私有制范围
- 个体经济
- 私营经济
- 外资经济和混合所有制经济中的非公有成分
- 坚决落实**两个毫不动摇**
    - 毫不动摇巩固和发展公有制经济
    - 毫不动摇鼓励、支持、引导非公有制经济发展
- **优化营商环境**作为抓产业发展最紧迫的任务

#### 分配制度演变 ####
第三次分配
- 基于自愿原则，在道德力量、精神力量、文化因素，价值追求等因素的综合影响下，实现社会财富资源在单向度流动
初次分配
- 强调在社会主义市场经济条件下，按照劳动力、资本、土地、技术等各种生产要素在国民收入中做出的贡献的大小比例所进行的分配
再次分配
- 政府根据法律法规，在初次分配的基础上，利用征收税收和政府非税收入的方式对各个收入主体所获得的收入进行再一次分配的过程

第三次分配出现的新形式
- 网络慈善模式
- 以购代捐模式
- 专业志愿服务模式
- 时间银行模式



收入代际传递
1. 代际传递：父辈收入、地位、职业、教育背景、生活环境等因素与子代相应因素之间存在很强的正相关；
2. 特征：代际传递越强，说明子代要突破父辈约束向上流动的可能性越小。


国有企业的双重属性
1. 企业属性（一般属性）
2. 作为社会主义公有制主要实现形式的**所有制属性**

产业政策
1. 产业结构政策
    - 主导产业选择政策
    - 弱小产业扶植政策
    - 衰退产业调整政策
2. 产业组织政策
    - 反垄断和不正当竞争政策
    - 中小企业政策
3. 产业发展政策
    - 产业技术政策
    - 产业布局政策
    - 产业外贸政策
    - 产业财政金融政策
    - 产业可持续发展政策
