### 第一章 导论 ###
1. 政治经济学的研究对象
    - 研究**社会生产关系**的科学$\qquad\qquad\qquad\qquad$生产关系：人们在生产过程中结成的关系
2. 物质资料的生产是人类社会存在和发展的基础，是政治经济学研究的出发点
    - 生产关系存在于物质资料的生产中
    - 物质资料的生产过程：人类通过自身的劳动，征服自然、改造自然，将自然界原有的自然物质改变成为适合于人类需要的物质财富的活动
    - 产生两大关系：
        - 人与物的关系——生产力
        - 人与人的关系——生产关系
    - 物质资料的生产三要素
        - 劳动者
        - 劳动对象
        - 劳动资料（包含**劳动工具**：代表生产力的发展水平）
    - 广义的生产关系包括**生产**、**分配**、**交换**、**消费**
    - 生产力：人们征服自然、改造自然的能力
        - 三要素：人、劳动资料、劳动对象
        - 劳动资料和劳动对象统称生产资料
        - 科学技术是第一生产力
3. **生产关系**——人类社会最本质最根本的关系，决定人类社会的性质，表现为人与人的物质利益关系
    - 生产、分配、交换、消费四大关系的总和
    - 三个方面的内容
        - 生产资料的所有制关系——最基础、最核心，决定生产关系中的其他两个关系
        - 人们在生产过程中的形成的地位及相互关系——由生产资料所有制关系决定
        - 产品的分配关系——由前两种关系决定
4. 生产力与生产关系的关系
    - 构成人类社会的生产方式
    - 生产力决定生产关系
    - 生产关系反作用于生产力
5. 经济基础与上层建筑
    - 经济基础：一定的生产关系的总和，经济基础是由生产关系构成的，
    - 上层建筑：建立在一定经济基础之上的政治法律制度和社会意识形态
    - 关系：对立统一，上层建筑适应经济基础时，上层建筑促进经济基础，上层建筑不适应经济基础时，上层建筑阻碍经济基础
 


### 第二章 政治经济学的研究方法 ###
1. 唯物辩证法（矛盾的方法）
   - 质量互变规律
   - 对立统一规律
   - 否定之否定规律
   - 矛盾是事物之间的对立统一，任何事物都是由矛盾构成的，事物的本质是由矛盾的主要方面决定的
   - 资本主义条件下最普遍的经济现象是商品
2. 抽象法
    - 从具体到抽象的研究方法
    - 从抽象到具体的叙述方法
3. 历史与逻辑一致的方法
    - 逻辑方法是在研究社会经济现象时所采用的思维推理法
    - 历史方法是在研究社会经济现象时按照历史发展过程来进行研究的方法

### 第二章 商品与货币 ###
1. 商品
   - 劳动价值论的核心：劳动创造价值
   1. 商品
        - 是用来交换的劳动产品
        - 两种形态：实物形态和价值形态
        - **使用价值**：商品的有用性，反映人与物的关系，是商品的自然属性，指商品体本身
        - **价值**：凝结在商品中的**无差别的一般人类劳动**，商品的本质属性
        - 要想拥有使用价值，就必须让渡价值

3. 货币
4. 价值物
5. 劳动的二重性：具体劳动与抽象劳动
    1. 具体劳动：劳动的具体方式
    2. 抽象劳动：劳动者在生产劳动中所耗费的各种具体形式的、无差别的、一般意义上的人类劳动，抽象劳动构成商品价值的实体
6. 私有制商品经济的基本矛盾：私人劳动与社会劳动的矛盾
    - 私人劳动：劳动的私有性，由生产资料私有制决定，
