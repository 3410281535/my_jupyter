### 第一章 导论 ###
1. 政治经济学的研究对象
    - 研究**社会生产关系**的科学$\qquad\qquad\qquad\qquad$生产关系：人们在生产过程中结成的关系
2. 物质资料的生产是人类社会存在和发展的基础，是政治经济学研究的出发点
    - 生产关系存在于物质资料的生产中
    - 物质资料的生产过程：人类通过自身的劳动，征服自然、改造自然，将自然界原有的自然物质改变成为适合于人类需要的物质财富的活动
    - 产生两大关系：
        - 人与物的关系——生产力
        - 人与人的关系——生产关系
    - 物质资料的生产三要素
        - 劳动者
        - 劳动对象
        - 劳动资料（包含**劳动工具**：代表生产力的发展水平）
    - 广义的生产关系包括**生产**、**分配**、**交换**、**消费**
    - 生产力：人们征服自然、改造自然的能力
        - 三要素：人、劳动资料、劳动对象
        - 劳动资料和劳动对象统称生产资料
        - 科学技术是第一生产力
3. **生产关系**——人类社会最本质最根本的关系，决定人类社会的性质，表现为人与人的物质利益关系
    - 生产、分配、交换、消费四大关系的总和
    - 三个方面的内容
        - 生产资料的所有制关系——最基础、最核心，决定生产关系中的其他两个关系
        - 人们在生产过程中的形成的地位及相互关系——由生产资料所有制关系决定
        - 产品的分配关系——由前两种关系决定
4. 生产力与生产关系的关系
    - 构成人类社会的生产方式
    - 生产力决定生产关系
    - 生产关系反作用于生产力
5. 经济基础与上层建筑
    - 经济基础：一定的生产关系的总和，经济基础是由生产关系构成的，
    - 上层建筑：建立在一定经济基础之上的政治法律制度和社会意识形态
    - 关系：对立统一，上层建筑适应经济基础时，上层建筑促进经济基础，上层建筑不适应经济基础时，上层建筑阻碍经济基础
