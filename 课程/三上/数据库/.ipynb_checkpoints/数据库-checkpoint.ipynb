**数据库管理系统**由一个互相关联的数据的集合和一组用以访问这些数据的程序组成。
<BR><BR>

数据库系统用于管理以下数据集合：
1. 高价值；
2. 相对庞大；
3. 多用户、并发访问。
<BR><BR>

早期数据库应用程序直接*构建在**文件系统**上*，导致以下**弊端**：
1. **数据冗余**（相同信息在几个地方重复储存）和**不一致性**（同一数据的不同副本不一致）；
2. **数据访问困难**（传统的文件处理环境不支持以一种方便而高效的方式去获取所需数据）；
3. **数据孤立**（数据分散在不同文件中，编写新的应用程序来检索困难）；
4. **完整性问题**（数据的值必须满足某些特定类型的*一致性约束*，添加新约束时修改程序困难）；
5. **原子性问题**（某一组操作必须是原子的，这在传统文件处理系统中很困难）；
6. **并发访问异常**；
7. **安全性问题**（难向用户提供*部分而不是全部*数据，以实现安全性约束）。
<BR>

数据库系统为上述所有问题提供解决方案。
<BR><BR>

数据库系统的一个**主要目的**：*向用户提供抽象的数据视图*。
<BR><BR>
