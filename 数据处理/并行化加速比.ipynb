### Amadahl定律
#### <font face="黑体">刻画了并行化的加速比
##### <font face="黑体"> 处于无法并行化的部分时其他节点需等待

<font size="3">*$Speedup_{Amadahl}$* = *$\frac{T_{sequencial}}{T_{parallel}}$* = $\frac1{(1 - f) +\frac{f}{m}}$

f : 一个任务中可并行计算的部分所占的比例  
m : 并行处理的节点个数

```python
import matplotlib.pyplot as plt
import numpy as np

# 参数设置
k = [0.1, 0.3, 0.5, 0.7, 0.9]
m = np.logspace(0, 15, num=100, base=2)  # 生成从2^0到2^15的对数均匀分布的100个点

# 创建图形
plt.figure(figsize=(10, 7))

# 计算并绘制每个 f 值的加速比
for f in k:
    Speedup = [1 / (1 - f + (f / i)) for i in m]
    plt.plot(m, Speedup, label=f'f = {f}')

# 添加标题和标签
plt.title("Amdahl")
plt.xlabel("Number of processors (m)")
plt.ylabel("Speedup")

# 设置横轴为对数刻度
plt.xscale('log', base=2)

# 设置横轴刻度
plt.xticks(ticks=[pow(2, i) for i in range(16)], labels=[f'{pow(2, i)}' for i in range(16)])

# 显示网格线和图例
plt.grid(True, which='both', linestyle='--', color='gray')
plt.legend()

# 显示图形
plt.show()
```

### Gustafson定律
#### <font face="黑体"> 在Amdahl定律基础上，处于无法并行化的环节时可以切换到其他任务


*$Speedup_{Gustafson}$* = *$\frac{T_{sequencial}}{T_{parallel}}$* = $\frac{1 / [(1 - f) + f]}{1 / [(1 - f) + mf]}$  = $(1 - f) + mf$ = $1 + (m - 1)f$

```python
m = range(0, 121, 20)
k = [0.1, 0.3, 0.5, 0.7, 0.9]

# 计算加速比
speedup_gustafson = []
for f in k:
    s = [1 + (mi - 1) * f for mi in m]
    speedup_gustafson.append(s)

# 创建图形
plt.figure(figsize=(10, 6))

# 绘制每个 f 值的加速比
for i, f_value in enumerate(k):
    plt.plot(m, speedup_gustafson[i], label=f'f = {f_value}')

# 添加标题和标签
plt.title("Gustafson")
plt.xlabel("Number of processors (m)")
plt.ylabel("Speedup")

# 添加图例
plt.legend()

# 显示网格
plt.grid(True)

# 显示图形
plt.show()
```
